```python
#lab: bmi 계산하기
import numpy as np
heights = [1.83, 1.76, 1.69, 1.86, 1.77, 1.73]
weights = [86, 74, 59, 95, 80, 68]
h_m = np.array(heights)
w_m = np.array(weights)
bmi = w_m/(h_m**2)
print(bmi)
```

```python
#lab: 잡음이 들어간 직선 그리기
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(100)
pure = np.linspace(1,10,100)
noise = np.random.randn(100)
line = pure + noise
plt.plot(line)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
g1 = np.random.normal(0,1,size = 6000)
g2 = np.random.normal(10,2,size = 6000)
plt.hist(g1)
plt.hist(g2)
plt.show()
```

```python
#lab: 싸인 함수 그리기
pi = 3.14
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-6,6,100)
s1 = np.linspace(-2*pi,2*pi,100)
sin1 = np.sin(s1)
sin2 = 3*sin1
plt.plot(x,sin1,x,sin2)
plt.show()
```

```python
#lab: mse 오차 계산하기
import numpy as np
ypred = np.array([1,0,0,0,0])
y = np.array([0,1,0,0,0])
mse = ((ypred-y)**2)/5
MSE = sum(mse)
print(MSE)
```

```python
#LAB: 직원들의 월급 인상하기
import numpy as np
salary = np.array([220,250,230])
salary1 = salary+100
print(salary1)
salary2 = salary * 2
print(salary2[salary2>450])
```
